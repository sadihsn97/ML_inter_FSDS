### Q-2. Imagine you have a dataset where you have different features like Age ,Gender , Height , Weight , BMI , and Blood Pressure and you have to classify the people into different classes like Normal , Overweight , Obesity , Underweight , and Extreme Obesity by using any 4 different classification algorithms. Now you have to build a model which can classify people into different classes.

Dataset:https://www.kaggle.com/datasets/ankurbajaj9/obesity-levels

```python

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv('/content/ObesityDataSet_raw_and_data_sinthetic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
## segregate numerical and categorical columns

numerical_columns=df.columns[df.dtypes!='object']
categorical_columns=df.columns[df.dtypes=='object']
print("Numerical columns:",numerical_columns)
print('Categorical Columns:',categorical_columns)
```

```python


df.nunique()

```

```python


from sklearn.preprocessing import LabelEncoder

```

```python


le = LabelEncoder()
df['NObeyesdad'] = le.fit_transform(df['NObeyesdad'])
df['SMOKE'] = le.fit_transform(df['SMOKE'])
df['MTRANS'] = le.fit_transform(df['MTRANS'])
df['family_history_with_overweight'] = le.fit_transform(df['family_history_with_overweight'])
df['CAEC'] = le.fit_transform(df['CAEC'])
df['CALC'] = le.fit_transform(df['CALC'])
df['FAVC'] = le.fit_transform(df['FAVC'])
df['SCC'] = le.fit_transform(df['SCC'])
df.head()

```

```python
target = df['NObeyesdad']
df.drop('NObeyesdad',axis = 'columns',inplace = True)
df.head()
```

```python


sex_columns  = pd.get_dummies(df['Gender'])

```

```python


df.drop(['Gender'],axis = 1, inplace = True)
df = df.join([sex_columns])
df.head()

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df,target,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
DecisionTreePipeline = Pipeline([('Decision_Tree',DecisionTreeClassifier())])
RandomForestClassifierPipeline = Pipeline([('RF_Classifier',RandomForestClassifier())])
```

```python
mypipeline = [DecisionTreePipeline,RandomForestClassifierPipeline]
```

```python


accuracy = 0.0
classifier = 0
pipeline = ""
```

```python
X_train
```

```python
pipelinedict = {0:'DecisionTreeClassifier',1:'RandomForestClassifier'}
for mypipe in mypipeline:
    mypipe.fit(X_train,y_train)
```

```python


for i, model in enumerate(mypipeline):
    print('{} Test Score = {}'.format(pipelinedict[i],model.score(X_test,y_test)))

```
